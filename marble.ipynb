```python
import requests
import json

url = "https://www.courtlistener.com/api/rest/v3/dockets/?court__id=gactapp&order_by=-date_modified"
data = []

# Fetch data from the first page
response = requests.get(url)
response_data = response.json()
print(response_data)
data.extend(response_data['results'])

# Fetch data from the next 2 pages
for _ in range(2):
    next_url = response_data['next']
    if next_url:
        response = requests.get(next_url)
        response_data = response.json()
        data.extend(response_data['results'])
    else:
        break

# Save the data to a JSON file
with open('dockets_data_ga_ct.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)

print("Data saved to dockets_data_ill_ct.json")
```

```python
# Read the JSON file
with open('dockets_data_ga_ct.json', 'r') as json_file:
    data = json.load(json_file)

data = [item['clusters'][0] for item in data if item['clusters'][0]]
final_dict = []
for url in data:
    response = requests.get(url)
    response_data = response.json()
    print(response_data)
    final_dict.append({'id':response_data['id'],'opinion_url': f"https://www.courtlistener.com{response_data['absolute_url']}", 'case_name':response_data['case_name']})
print(final_dict)

# Save the data to a JSON file
with open('opinion_data_ga_ct.json', 'w') as json_file:
    json.dump(final_dict, json_file, indent=4)
```
